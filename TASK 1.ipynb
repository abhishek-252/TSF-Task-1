# <u>TASK 1</u> - PREDICTION USING SUPERVISED ML
In this regression task we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied. This is a simple linear regression task as it involves just two variables.

### 1. Making necessary imports 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### 2. Our data 

```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
```

### 3.Understanding the dataset

```python
data.head(10)
```

### 4. Checking for null values

```python
data.isnull().sum()
```

```python
len(data)
```

### 5. Visualizing our data

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Hours',y='Scores',data=data)
plt.xlabel('No. of hours studied')
plt.ylabel('Score obtained')
plt.title('No. of hours dedicated vs Scores obtained');
```

### 6. Preparing data

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

### 7. Splitting data into training and testing data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

### 8. Importing multiple predictors
*we'll make our predictions and then choose the best performing model based on the degree of error they produce.*

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from catboost import CatBoostRegressor
```

```python
linreg = LinearRegression()
ranfor = RandomForestRegressor()
cat = CatBoostRegressor()
```

### 9. Training our models

```python
linreg.fit(X_train,y_train)
ranfor.fit(X_train,y_train)
cat.fit(X_train,y_train)
```

### 10. Making the predictions

```python
lin_preds = linreg.predict(X_test)
ran_preds = ranfor.predict(X_test)
cat_preds = cat.predict(X_test)
```

### 11. Calculating errors produced by each predictor 

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, lin_preds)) 
```

```python
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, ran_preds)) 
```

```python
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, cat_preds)) 
```

---
## <u>*NOTE*</u> : WE'LL GO AHEAD WITH LINEAR REGRESSION MODEL AS IT HAS THE LOWEST MAE
---

### 12. Visualizing our predicted regression line.

```python
line = linreg.coef_*X+linreg.intercept_
plt.figure(figsize=(10,6))
plt.scatter(X, y)
plt.plot(X, line,color='red')
plt.xlabel('No. of hours studied')
plt.ylabel('Score obtained')
plt.title('No. of hours dedicated vs Scores obtained');
```

### 13. Predicting score for student who studied 9.25 hrs/day 

```python
print('Score of a student who studied for 9.25hrs/day is :',linreg.predict([[9.25]])[0])
```
